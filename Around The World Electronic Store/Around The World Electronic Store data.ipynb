```python
import pandas as pd
```

```python
df=pd.read_csv("d:\Z-Games\Orders.csv",skiprows=4)
```

```python
df.head()
```

```python
df['Discount']=df['Discount'].fillna(int(0))
df=df.dropna()
```

```python
df.isnull().sum()
```

```python

df["Phone Number"]=df["Phone Number"].astype(str).str.replace("Tel:"," ")
df["Phone Number"]=df["Phone Number"].astype(int)
df["Order ID"]=df["Order ID"].astype(int)
df["Quantity"]=df["Quantity"].astype(int)
df["total unit sold"]=df['Quantity']*df["Unit Price"]

```

```python
df.head()
```

```python
df["Customer First Name"]=df["Customer First Name"].str.lower().str.capitalize()
df["Customer Last Name"]=df["Customer Last Name"].str.lower().str.capitalize()
df["Order Date"]=pd.to_datetime(df["Order Date"],dayfirst=True,errors="coerce")
```

```python
df.to_csv("Around The World Phone Store.csv",index=False)
```
